This example shows how to use the PyVLM class to generate the wing planform of a Pilatus PC12 airplane, apply the VLM theory and obtain CL(aplha) and CD(alpha).

The right wing will be defined using the following nomenclature while the left wing will be created as its specular image:                           
![pilatusPC12](pc12.png)

The geometry is defined as follows:

```python
import numpy as np
import matplotlib.pyplot as plt

from pyvlm.vlm import PyVLM

pilatusPC12 = PyVLM()

# Geometry parameters
A = np.array([0, 1.03])  # root chord leading edge position
B = np.array([.414, 8.14])  # tip chord leading edge position
chord_1 = 2.15
chord_2 = 1.24

leading_edges_position = [A, B]
chord_length = [chord_1, chord_2]  # root, tip
```

Then the mesh is generated using the "add_wing" method of the PyVLM class, and visualize using the "check_mesh" method:

```python
n, m = 4, 3  # number of panels (chordwise, spanwise)
pilatusPC12.add_wing(leading_edges_position, chord_length, n, m)
pilatusPC12.check_mesh(printt=False, plott=True)
```

Finally, the values for CL, CD and alpha are obtained as lists returned from the method "aerodyn_forces_coeff":

```python
alpha, CL, CD = pilatusPC12.aerodyn_forces_coeff()

plt.style.use('ggplot')

plt.subplot(1, 2, 1)
plt.title("CL")
plt.xlabel("alpha")
plt.plot(alpha, CL)

plt.subplot(1, 2, 2)
plt.title("CD")
plt.xlabel("alpha")
plt.plot(alpha, CD)

plt.show()
```
